```python
import pandas as pd;
from matplotlib import pyplot as plt;
import seaborn as sns;
%matplotlib inline
import warnings;
warnings.filterwarnings('ignore');
```

```python
data = pd.read_csv('Homework-Dataset.csv',encoding='latin1')
```

```python
data.head()
```

```python
data.columns = ['Day', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'Budget', 'Gross',
       'IMDbRating', 'MovieLensRating', 'Overseas', 'OverseasPercent',
       'Profit', 'ProfitPercent', 'Runtime', 'US',
       'GrossPercentUS']
```

```python
#plot graph
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(20,8.27)})

s = sns.swarmplot(data=data, x='Genre', y='GrossPercentUS', hue='Studio')
b = sns.boxplot(data=data, x='Genre', y='GrossPercentUS', color='lightgrey')
box = b.get_position()
b.set_position([box.x0, box.y0, box.width * 0.85, box.height])

b.legend(bbox_to_anchor=(1.20, 1), ncol=1, frameon=False, fancybox=False, shadow=False)

plt.title("Domestic Gross % by Genre", fontsize=40)
plt.xlabel("Genre", fontsize=40)
plt.ylabel("Gross % US", fontsize=40)
```
